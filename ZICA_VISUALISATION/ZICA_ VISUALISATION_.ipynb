```python
import numpy as np # linear algebra

import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import os
%matplotlib inline
from subprocess import check_output
```

```python
df = pd.read_csv("cdc_zika.csv",parse_dates=['report_date'],infer_datetime_format=True,index_col=0)
df.head(5)
```

```python
#to 30 locations of highly infected area

df.location.value_counts()[:30].plot(kind='bar', figsize=(12, 7))

plt.title("Number of locations reported - Top 30")
```

```python

# Graph on the basis of confirmed sympomts


symptoms = ['confirmed_fever',
       'confirmed_acute_fever', 'confirmed_arthralgia',
       'confirmed_arthritis', 'confirmed_rash', 'confirmed_conjunctivitis',
       'confirmed_eyepain', 'confirmed_headache', 'confirmed_malaise']
fig = plt.figure(figsize=(13,13))
for symptom in symptoms:
    df[df.data_field == symptom].value.plot()
plt.legend(symptoms, loc='best')
plt.title('Understanding symptoms of zika virus')
```

```python
# distribution on the basis country-state-territory-department- region -city

df.location_type.value_counts()[:30].plot(kind='bar', figsize=(12, 7))
```

```python
df.data_field.value_counts()[:60].plot(kind='bar',color='c', figsize=(12, 7))
```

```python
df.data_field.value_counts()[4:60].plot(kind='bar',color='c', figsize=(12, 7))
```

```python
df.data_field.value_counts()[60:100].plot(kind='bar',color='c', figsize=(12, 7))
```

```python
 ## data of confirmed male VS confirmed Female cases  


df[df.data_field == "confirmed_male"].value.plot()

df[df.data_field == "confirmed_female"].value.plot().legend(("Male","Female"),loc="best")

plt.title("Confirmed Male vs Female cases")
```

```python
## graph distibution on the bases of age

age_groups = ['confirmed_age_20to39', 'confirmed_age_40to59',
       'confirmed_age_over59', 'confirmed_age_under20', 'confirmed_age_unk']
fig = plt.figure(figsize=(10,6))
for age in age_groups:
    df[df.data_field == age].value.plot()
plt.legend(age_groups, loc='best')
plt.title('Understanding symptoms of zika virus')
```
